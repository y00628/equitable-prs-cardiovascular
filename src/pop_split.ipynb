```python
import pandas as pd
```

```python
pop_data = pd.read_csv('../data/integrated_call_samples_v3.20130502.ALL.panel', sep='\t')
pop_data
```

```python
pop_data['super_pop'].value_counts()
```

```python
pops = pop_data['super_pop'].unique()
pops
```

```python
for pop in pops:
    pop_data[pop_data['super_pop']==pop][['sample']].to_csv(f'../data/sample_populations/{pop}', sep='\t', index=False, header=None)
```

```python
# List of chromosomes (you can add 'X' and 'Y' if you want to include them)
chromosomes = [str(i) for i in range(1, 23)]  # For chromosomes 1 to 22
# chromosomes = [str(i) for i in range(1, 23)] + ['X', 'Y']  # Optionally add X and Y

# Base paths for VCF and output files
vcf_base_path = "../data/1000_genomes_vcf/ALL.chr"
chr_base_path = "../data/1000_genomes_bbf/chr"

# Loop through each chromosome and run the PLINK command
for pop in pops:
    for chrom in chromosomes:
        chr_file = f"{chr_base_path}{chrom}"
        
        # Construct the PLINK command
        plink_command = f"plink2 --bfile {chr_file} --keep ../data/sample_populations/{pop} --make-bed --out ../data/1000_genomes_split/{pop}_chr{chrom}"
        
        # Execute the command (using the ! operator to run the shell command in Jupyter)
        print(f"Running command for population {pop}, chromosome {chrom}:")
        print(plink_command)
        !{plink_command}


```
